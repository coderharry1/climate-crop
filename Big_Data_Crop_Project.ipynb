# Modeling 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ------------------ 1. Train / validation / test split ------------------
```

```python
y = df_cleaned['Yield_tonnes_per_ha']
X = df_cleaned.drop('Yield_tonnes_per_ha', axis=1)
```

```python
categorical_cols = ['Country', 'Crop']
numeric_cols = ['Year']
```

```python
preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
    ('num', StandardScaler(), numeric_cols)
])
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
# Split remaining 85% into train (70%) and validation (15%)
val_size = 0.15 / 0.85  
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=val_size, random_state=42)
```

```python
print('Train shape:', X_train.shape)
print('Validation shape:', X_val.shape)
print('Test shape:', X_test.shape)
```

```python
# ------------------ 2. Baseline model: Linear Regression ------------------
```

```python
lin_model = Pipeline([
    ('preprocess', preprocessor),
    ('model', LinearRegression())
])
```

```python
lin_model.fit(X_train, y_train)
```

```python
val_pred_lin = lin_model.predict(X_val)
lin_rmse_val = mean_squared_error(y_val, val_pred_lin) ** 0.5
lin_r2_val = r2_score(y_val, val_pred_lin)
```

```python
print('Linear Regression Validation RMSE:', lin_rmse_val)
print('Linear Regression Validation R2:', lin_r2_val)
```

```python
# ------------------ 3. complex models ------------------
```

```python
rf_model = Pipeline([
    ('preprocess', preprocessor),
    ('model', RandomForestRegressor(random_state=42))
])
```

```python
gb_model = Pipeline([
    ('preprocess', preprocessor),
    ('model', GradientBoostingRegressor(random_state=42))
])
```

```python
#hyperparameters
rf_params = {
    'model__n_estimators': [200, 400, 600],
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_leaf': [1, 2, 4]
}
```

```python
gb_params = {
    'model__n_estimators': [100, 200, 400],
    'model__learning_rate': [0.05, 0.1, 0.2],
    'model__max_depth': [2, 3, 4]
}
```

```python
rf_search = RandomizedSearchCV(rf_model, rf_params, n_iter=10, cv=3, scoring='neg_root_mean_squared_error', random_state=42, n_jobs=-1)
rf_search.fit(X_train, y_train)

gb_search = RandomizedSearchCV(gb_model, gb_params, n_iter=10, cv=3, scoring='neg_root_mean_squared_error', random_state=42, n_jobs=-1)
gb_search.fit(X_train, y_train)
```

```python
# Evaluating on validation
rf_val_pred = rf_search.best_estimator_.predict(X_val)
gb_val_pred = gb_search.best_estimator_.predict(X_val)
```

```python
rf_rmse_val = mean_squared_error(y_val, rf_val_pred) ** 0.5
gb_rmse_val = mean_squared_error(y_val, gb_val_pred) ** 0.5
rf_r2_val = r2_score(y_val, rf_val_pred)
gb_r2_val = r2_score(y_val, gb_val_pred)
```

```python
metrics = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest (tuned)', 'Gradient Boosting (tuned)'],
    'RMSE_val': [lin_rmse_val, rf_rmse_val, gb_rmse_val],
    'R2_val': [lin_r2_val, rf_r2_val, gb_r2_val]
})
```

```python
print(metrics)
```

```python
# ------------------ 4. Select best model and evaluate on test ------------------
```

```python
best_model = rf_search.best_estimator_ if rf_rmse_val < gb_rmse_val else gb_search.best_estimator_

best_model.fit(pd.concat([X_train, X_val]), pd.concat([y_train, y_val]))

test_pred = best_model.predict(X_test)
```

```python
test_rmse = mean_squared_error(y_test, test_pred) ** 0.5
test_r2 = r2_score(y_test, test_pred)
print('Best Model on Test RMSE:', test_rmse)
print('Best Model on Test R2:', test_r2)
```

```python
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

X_val_sampled = X_val.sample(n=100, random_state=42)
y_val_sampled = y_val.loc[X_val_sampled.index]

X_val_transformed = best_pipe.named_steps['preprocess'].transform(X_val_sampled)
if hasattr(X_val_transformed, 'toarray'):
    X_val_transformed = X_val_transformed.toarray()

from sklearn.pipeline import Pipeline
model = best_pipe.named_steps['model'] if isinstance(best_pipe, Pipeline) else best_pipe

result = permutation_importance(
    model,
    X_val_transformed,
    y_val_sampled,
    n_repeats=5,
    random_state=42,
    scoring='r2'
)

encoder = best_pipe.named_steps['preprocess'].named_transformers_['cat']
onehot_feature_names = encoder.get_feature_names_out(['Country', 'Crop'])
all_feature_names = np.concatenate([onehot_feature_names, ['Year']])

perm_df = pd.DataFrame({
    'feature': all_feature_names,
    'importance_mean': result.importances_mean,
    'importance_std': result.importances_std
}).sort_values('importance_mean', ascending=True).tail(10)

sns.set(style="whitegrid", context="notebook")
colors = sns.color_palette("rocket", n_colors=len(perm_df))

fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(
    data=perm_df,
    x='importance_mean',
    y='feature',
    palette=colors,
    xerr=perm_df['importance_std'],
    edgecolor='black',
    ax=ax
)

for patch in ax.patches:
    width = patch.get_width()
    y = patch.get_y() + patch.get_height() / 2
    ax.text(
        width + 0.02,
        y,
        f'{width:.2f}',
        va='center',
        fontsize=10,
        color='black'
    )

ax.set_title('🔍 Permutation Feature Importance (Top 10)', fontsize=16, weight='bold', loc='left')
ax.set_xlabel('Mean Importance (Decrease in R²)', fontsize=13)
ax.set_ylabel('Feature', fontsize=13)
ax.tick_params(labelsize=11)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.grid(True, linestyle='--', alpha=0.6)
ax.yaxis.grid(False)

plt.tight_layout()
plt.savefig('/Users/harish/Downloads/Figure4_PermutationImportance_Top10_Styled.png', dpi=300)
plt.show()

```
