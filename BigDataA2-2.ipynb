# Filter for Yield

```python
df_yield = df[df['Element'] == 'Yield'].copy()
print(df_yield.shape)
print(df_yield['Element'].unique())
```

# Inspecting Available Countries

```python
print(df_yield['Area'].unique())
```

```python
study_countries = ['Australia', 'Brazil', 'India', 'Nigeria', 'South Africa']
```

# Use a yield dataframe filter.

```python
df_yield = df_yield[df_yield['Area'].isin(study_countries)]
print(df_yield['Area'].unique())
```

```python
print(df_yield.shape)
```

# Identify Year Columns

FAOSTAT has columns like:

Y1961
Y2021F
Y2021N

```python
import re
year_columns = [col for col in df_yield.columns if re.match(r'^Y\d{4}$', col)]
```

```python
print(year_columns[:10])
```

# Melt (pivot) from wide to long

```python
df_long = df_yield.melt(
    id_vars=['Area', 'Item', 'Element', 'Unit'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Value'
) 
```

```python
print(df_long.shape)
print(df_long.head())
```

# Cleaning Year Column

```python
df_long['Year'] = df_long['Year'].str.extract(r'(\d{4})').astype(int)
print(df_long['Year'].unique()[:10])
```

# Drop Missing Yield Values

```python
df_long = df_long.dropna(subset=['Value'])
print(df_long.shape)
```

# Confirming if no missing values 

```python
df_long = df_long.dropna(subset=['Value'])
print(df_long.shape)
```

```python
df_long = df_long.dropna(subset=['Value'])
print(df_long.shape)
```

```python
df_long.head()
```

```python
# 1️⃣ Removing rows with Missing Values
df_long = df_long.dropna(subset=['Value'])

# 2️⃣ Convert kg/ha to tonnes/ha
df_long['Yield_tonnes_per_ha'] = df_long['Value'] / 1000

df_cleaned = df_long[['Area', 'Item', 'Year', 'Yield_tonnes_per_ha']].copy()

df_cleaned.rename(columns={
    'Area': 'Country',
    'Item': 'Crop'
}, inplace=True)

print(df_cleaned.head())
print(df_cleaned.shape)
```

```python
# Saving 
output_path = '/Users/harish/Downloads/Cleaned_FAOSTAT_Yield.csv'
df_cleaned.to_csv(output_path, index=False)

# Confirming if it is saved
print(f"✅ File saved to: {output_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
file_path = '/Users/harish/Downloads/Cleaned_FAOSTAT_Yield.csv'
df = pd.read_csv(file_path)

print(df.shape)
print(df.head())
```

```python
print(df.describe())
```

```python
country_stats = df.groupby('Country')['Yield_tonnes_per_ha'].describe()
print(country_stats)
```

```python
sns.set(style="whitegrid", context="talk")

plt.figure(figsize=(12,7))
sns.violinplot(
    data=df,
    x='Country',
    y='Yield_tonnes_per_ha',
    inner=None,
    palette='pastel'
)
sns.boxplot(
    data=df,
    x='Country',
    y='Yield_tonnes_per_ha',
    width=0.15,
    color='k'
)
plt.ylim(0, 40)
plt.title('Violin + Boxplot of Crop Yields (tonnes/ha) by Country', fontsize=18, weight='bold')
plt.ylabel('Yield (tonnes/ha)', fontsize=14)
plt.xlabel('Country', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig('/Users/harish/Downloads/Figure1_Violin_Boxplot_Yield_Country.png', dpi=300)
plt.show()
```

# Trends in Mean Yield by Country Over Time

```python
plt.figure(figsize=(12,6))
sns.set(style="whitegrid", context="talk")
sns.lineplot(
    data=df,
    x='Year',
    y='Yield_tonnes_per_ha',
    hue='Country',
    estimator='mean',
    palette='Set2'
)
plt.title('Mean Crop Yield Trends Over Time by Country')
plt.ylabel('Mean Yield (tonnes/ha)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.tight_layout()
plt.savefig('/Users/harish/Downloads/Figure2_Yield_Trends_Country.png', dpi=300)
plt.show()
```

# Crop vs. Country Average Yield Heatmap

```python
pivot_table = df_cleaned.pivot_table(
    index='Crop',
    columns='Country',
    values='Yield_tonnes_per_ha',
    aggfunc='mean'
)

plt.figure(figsize=(14,16))
sns.heatmap(
    pivot_table,
    cmap='YlGnBu',
    cbar_kws={'label': 'Average Yield (tonnes/ha)'}
)
plt.title('Average Yield (tonnes/ha) by Crop and Country', fontsize=16, weight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Crop', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('/Users/harish/Downloads/Figure3_Heatmap_Yield_Crop_Country.png', dpi=300)
plt.show()
```
